## Question

Following are names of students who secured top 10 ranks in the aptitude exam.

1. Sachin
2. Sourav
3. Rahul
4. Yuvraj
5. Laxman
6. Viru
7. Venketesh
8. Javagal
9. Zahir
10. Mahendar



* Arrange the names in ascending order of number of letters contained in each.
* Find the number of repeating alphabets in each name.



```python
names = ["Sachin","Sourav","Rahul","Yuvraj","Laxman","Viru","Venketeshk","Javagal","Zahir","Mahendar"]

#Part one of the question
print("Names in ascending order of number of letters")
print("\t")
names.sort(key=len)
print(names)

#Part two of the question
print("\n")
print("Finding the number of repeating alphabets in each name")
for name in names:
    print("Name : ", name)
    count = {}
    for s in name:
        if s in count:
            count[s] += 1
        else:
            count[s] = 1
    #print(count)
    
    no_rep = 0
    for key in count:
        
        if count[key] > 1:
            no_rep = 1
            print (key, count[key], "times")
    if no_rep == 0:
        print("No repeatation")
        
    print("\t")
```
